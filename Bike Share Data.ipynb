- <h4> Data transformation and cleaning

We observe here that data type in start_station_id and end_station_id is not consistent in all the data sets. First three sets had start_station_id as int, while others were float64 and rest were objects. Similar was the case in the end_station_id.
we'll have to convert these columns into object.

```python
tripdata_202004.start_station_id = tripdata_202004.start_station_id.astype('str')
tripdata_202005.start_station_id = tripdata_202005.start_station_id.astype('str')
tripdata_202006.start_station_id = tripdata_202006.start_station_id.astype('str')
tripdata_202007.start_station_id = tripdata_202007.start_station_id.astype('str')
tripdata_202008.start_station_id = tripdata_202008.start_station_id.astype('str')
tripdata_202009.start_station_id = tripdata_202009.start_station_id.astype('str')
tripdata_202010.start_station_id = tripdata_202010.start_station_id.astype('str')
tripdata_202011.start_station_id = tripdata_202011.start_station_id.astype('str')
tripdata_202004.end_station_id = tripdata_202004.end_station_id.astype('str')
tripdata_202005.end_station_id = tripdata_202005.end_station_id.astype('str')
tripdata_202006.end_station_id = tripdata_202006.end_station_id.astype('str')
tripdata_202007.end_station_id = tripdata_202007.end_station_id.astype('str')
tripdata_202008.end_station_id = tripdata_202008.end_station_id.astype('str')
tripdata_202009.end_station_id = tripdata_202009.end_station_id.astype('str')
tripdata_202010.end_station_id = tripdata_202010.end_station_id.astype('str')
tripdata_202011.end_station_id = tripdata_202011.end_station_id.astype('str')

```

- <h4> Combine all the data sets into 1 frame

```python
all_trips_data = pd.concat([tripdata_202004,tripdata_202005,tripdata_202006,tripdata_202007,tripdata_202008,tripdata_202009,tripdata_202010,tripdata_202011,tripdata_202012,tripdata_202101,tripdata_202102,tripdata_202103,tripdata_202104,tripdata_202105,tripdata_202106,tripdata_202107],axis = 0)
```

```python
all_trips_data.shape
```

- <h4> started_at and ended_at should be in date time format.

```python
all_trips_data['started_at'] = pd.to_datetime(all_trips_data['started_at'])
```

```python
all_trips_data['ended_at'] = pd.to_datetime(all_trips_data['ended_at'])
```

```python
all_trips_data.head()
```

- <h4> Removing unnecessary columns

```python
all_trips_data = all_trips_data.drop(['start_lat','start_lng','end_lat','end_lng'],axis = 1)
```

```python
all_trips_data.head()
```

- <h4> Inserting necessary columns

```python
all_trips_data['start_date'] = all_trips_data['started_at'].dt.date
all_trips_data['start_time'] = all_trips_data['started_at'].dt.time
all_trips_data['end_date'] = all_trips_data['ended_at'].dt.date
all_trips_data['end_time'] = all_trips_data['ended_at'].dt.time
all_trips_data['month_year'] = all_trips_data['started_at'].dt.to_period('M')
```

```python
all_trips_data.head()
```

```python
# to find the duration of the trip
all_trips_data['duration'] = all_trips_data['ended_at'] - all_trips_data['started_at']
```

```python
all_trips_data.head()
```

```python
# converting the duration into minutes
all_trips_data['duration'] = pd.to_timedelta(all_trips_data['duration']).astype('timedelta64[m]').astype(int)
```
